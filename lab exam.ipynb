```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0][0]
intercept = model.intercept_[0]
print(f"Slope: {slope}, Intercept: {intercept}")

X_new = np.array([[0], [2]])
y_predict = model.predict(X_new)

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_new, y_predict, color='red', label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
X = np.hstack((X1, X2))
y = 4 + 3 * X1 + 5 * X2 + np.random.randn(100, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model_simple = LinearRegression()
model_simple.fit(X_train[:, 0].reshape(-1, 1), y_train)

model_multiple = LinearRegression()
model_multiple.fit(X_train, y_train)

y_simple_pred = model_simple.predict(X_test[:, 0].reshape(-1, 1))
y_multiple_pred = model_multiple.predict(X_test)

r_squared_simple = model_simple.score(X_test[:, 0].reshape(-1, 1), y_test)
r_squared_multiple = model_multiple.score(X_test, y_test)

print(f"Simple Linear Regression R-squared: {r_squared_simple}")
print(f"Multiple Linear Regression R-squared: {r_squared_multiple}")

plt.scatter(X_test[:, 0], y_test, color='blue', label='Data points (X1)')
plt.scatter(X_test[:, 1], y_test, color='orange', label='Data points (X2)')
plt.scatter(X_test[:, 0], y_simple_pred, color='red', label='Simple Model Prediction', marker='x')
plt.scatter(X_test[:, 0], y_multiple_pred, color='green', label='Multiple Model Prediction', marker='o')
plt.title('Multiple Linear Regression Comparison')
plt.xlabel('Independent Variables')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Generate synthetic data
np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
X = np.hstack((X1, X2))
y = 4 + 3 * X1 + 5 * X2 + np.random.randn(100, 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit the multiple linear regression model
model_multiple = LinearRegression()
model_multiple.fit(X_train, y_train)

# Make predictions
y_multiple_pred = model_multiple.predict(X_test)

# Calculate residuals
residuals = y_test - y_multiple_pred

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_multiple_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 100 * np.random.rand(100, 1)
X = np.hstack((X1, X2))
y = 4 + 3 * X1 + 5 * X2 + np.random.randn(100, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model_no_scaling = LinearRegression().fit(X_train, y_train)
y_pred_no_scaling = model_no_scaling.predict(X_test)
mse_no_scaling = mean_squared_error(y_test, y_pred_no_scaling)
r2_no_scaling = r2_score(y_test, y_pred_no_scaling)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_scaling = LinearRegression().fit(X_train_scaled, y_train)
y_pred_scaling = model_scaling.predict(X_test_scaled)
mse_scaling = mean_squared_error(y_test, y_pred_scaling)
r2_scaling = r2_score(y_test, y_pred_scaling)

print(f"Without Scaling: MSE = {mse_no_scaling:.2f}, R-squared = {r2_no_scaling:.2f}")
print(f"With Scaling: MSE = {mse_scaling:.2f}, R-squared = {r2_scaling:.2f}")
```
